## Data visualisation

```python
sns.scatterplot(data=df,x="Geography",y="EstimatedSalary",hue="Gender")
plt.xlabel("Geography")
plt.ylabel("EstimatedSalary")
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=df,x="Age",y="Balance",hue="Gender")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.legend()
plt.show()
```

```python
sns.barplot(data=df,x="Geography",y="EstimatedSalary",hue="Gender")
plt.xlabel("Geography")
plt.ylabel("EstimatedSalary")

plt.legend()
plt.show()
```

```python
france=(df['Geography']=="France").sum()
spain=(df['Geography']=="Spain").sum()
germany=(df['Geography']=="Germany").sum()

print(france,spain,germany)
```

```python
plt.pie(np.array([france,spain,germany]),labels=["France","Spain","Germany"],autopct='%.2f')
plt.legend(loc='upper right')
plt.show()
```

```python
sns.barplot(data=df,x="Geography",y="CreditScore", hue="Gender")
plt.xlabel("Geography")
plt.ylabel("CreditScore")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
